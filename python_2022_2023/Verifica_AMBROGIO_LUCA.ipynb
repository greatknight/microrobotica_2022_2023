```python
y=1
d = {}
for elemento in stringa_split2:
    
    parole = elemento[:-1].split("/")
    #print(parole[1])
    for i in parole:
        #print(i)
        par = i[:-1].split("/")
        d[i] = y
        y = y + 1

d
```

## ESERCIZIO 2
Crea un array di numpy che contenga tutti i numeri interi tra 2 e 1000 (estremi inclusi), ma nel quale tutti i numeri divisibili per 7 sono sostituiti da uno 0. Infine crea un array contenente tutti i numeri interi tra 2 e 1000 (estremi inclusi) ma privo di tutti i numeri divisibili per 7.

```python
import numpy as np
```

```python
a = np.arange(2,1001)
#b = np.arange(2,1001)
for i in range(len(a)):
    if a[i]%7==0:
        a[i]=0
b = np.array([a[i] for i in range(0,len(a)) if a[i]%7!=0])
print("a: ", a)
print("\n")
print("b: ", b)
```
