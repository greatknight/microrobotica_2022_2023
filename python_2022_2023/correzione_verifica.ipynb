## ESERCIZIO 1

```python
file = open("robots.txt")
righe = file.readlines()
file.close()
dizionario = {}
```

```python
for riga in righe[1:]:
    argomento = riga.split("/")[1]
    if(argomento in dizionario):
        dizionario[argomento] = dizionario[argomento] +1
    else:
        dizionario[argomento] = 1
    #print(argomento)
dizionario
```
