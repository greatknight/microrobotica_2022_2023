```python
num = input("Inserisci un numero: ")
n = int(num)
l = [2**i for i in range(0,n) if 2**i<n]
print(l)
```

```python
l = [i*i for i in range(0,200) if i*i<200 and i%2!=0]
print(l)
```

```python
l = [[y*i for y in range(1,11)] for i in range(1, 11)]
for z in range(10):
    print(l[z])
    print('\n')
```
