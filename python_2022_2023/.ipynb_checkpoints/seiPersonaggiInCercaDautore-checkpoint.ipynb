```python
from collections import Counter 
f=open("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt","r")
righe = f.readlines()
f.close()
```

```python
stringa = "".join(righe)
strlow = stringa.lower()
listaPun = [",", ";", ".", ":", "!", "?", "(", ")", "'", "-", "»", "«", "#", "¿", "ï"]
listaSenzPun  = [lettera for lettera in strlow if lettera not in listaPun]
stringaNoPun = "".join(listaSenzPun)
stringa_split2 = stringaNoPun.split()
count_items = Counter(stringa_split2) 
#print(count_items) 
```

```python
d = {}
i = 0

for parola in stringa_split2:
    #print(parola)
    if parola not in d:
        d[parola] = [stringa_split2[i+1] for i,par in enumerate(stringa_split2[:-1]) if parola==par and i <= len(stringa_split2)]
d
```

```python
def rimuoviPunteggiatura(testo,caratteri_da_eliminare):
    for carattere in caratteri_da_eliminare:
        testo.replace(carattere," ")
    return testo
caratteri_da_eliminare = [",",";",".",":","-","_","#","@","?","!","+","*","/","%","&","£","$","^","'"]
file = open("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt","r")
testo = file.readlines()
testo = [rimuoviPunteggiatura(riga,caratteri_da_eliminare) for riga in testo]
testo = [riga[:-1].split(" ")for riga in testo]
file.close()
dizionario = {}
lista_parole = []
for riga in testo:
    #riga[]
    for parola in riga:
        if parola!='':
            lista_parole.append(parola.lower())
        #scrivere in maniera più concisa
for parola in lista_parole:
    if parola not in dizionario:
        dizionario[parola] = 1
    else:
        dizionario[parola] += 1
```
